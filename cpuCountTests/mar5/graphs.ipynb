```python
import pandas as pd

file_names = ['mar5_8cpu.csv', 'mar5_16cpu.csv', 'mar5_36cpu.csv', 'mar5_76cpu.csv']

combined_data = pd.DataFrame()

for file_name in file_names:

    data = pd.read_csv(file_name)

    combined_data = pd.concat([combined_data, data], ignore_index=True)


combined_data.to_csv('mar5.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('mar5.csv')

shapes = df['data_shape'].unique()

for shape in shapes:
    shape_df = df[df['data_shape'] == shape]
    models = shape_df['model'].unique()

    for model in models:
        model_df = shape_df[df['model'] == model]

        cpu_counts = model_df['cpu_count'].unique()

        fig, ax = plt.subplots(figsize=(10, 10))

        for cpu_count in cpu_counts:
            cpu_df = model_df[model_df['cpu_count'] == cpu_count]

            # Group by 'vox_per_chunk' and calculate the mean and std of 'time'
            grouped_df = cpu_df.groupby('num_chunks')['time'].agg([np.mean, np.std]).reset_index()

            ax.errorbar(grouped_df['num_chunks'], grouped_df['mean'], yerr=grouped_df['std'], label=f'{cpu_count} CPUs', fmt='-o')

        ax.set_xscale('log')
        ax.set_ylim(bottom=0)

        ax.set_title(f'Model Type {model}, Shape {shape} (Average Time with Error Bars)')
        ax.set_xlabel('num_chunks (log scale)')
        ax.set_ylabel('Average Time')
        ax.legend()  # Add a legend
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('mar5.csv')

shapes = df['data_shape'].unique()

for shape in shapes:
    shape_df = df[df['data_shape'] == shape]
    models = shape_df['model'].unique()

    for model in models:
        model_df = shape_df[df['model'] == model]

        cpu_counts = model_df['cpu_count'].unique()

        fig, ax = plt.subplots(figsize=(10, 10))

        for cpu_count in cpu_counts:
            cpu_df = model_df[model_df['cpu_count'] == cpu_count]

            cpu_df['time_per_cpu'] = cpu_df['time'] * cpu_df['cpu_count']
            # Group by 'vox_per_chunk' and calculate the mean and std of 'time'
            grouped_df = cpu_df.groupby('num_chunks')['time_per_cpu'].agg([np.mean, np.std]).reset_index()

            ax.errorbar(grouped_df['num_chunks'], grouped_df['mean'], yerr=grouped_df['std'], label=f'{cpu_count} CPUs', fmt='-o')

        ax.set_xscale('log')
        ax.set_ylim(bottom=0)

        ax.set_title(f'Model Type {model}, Shape {shape} (Average Time per cpu with Error Bars)')
        ax.set_xlabel('num_chunks (log scale)')
        ax.set_ylabel('Average Time per cpu')
        ax.legend()  # Add a legend
        plt.show()
```
